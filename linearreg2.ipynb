```python
## data grouping
grouped_mean=df.groupby('Brands')['new Selling Price'].mean()
print(grouped_mean)
```

```python
## grouping with two category variables
grouped_mean=df.groupby(['Brands','Models'])['new Selling Price'].mean()
grouped_mean
```

```python
## aggregating multiple functins
groupped_agg=df.groupby(['Brands','Models'])['new Selling Price'].agg('mean','sum','count')
groupped_agg
```

```python
## merging and joining
df1=pd.DataFrame({'key':['A','B','C'],'value':[1,2,3]})
df2=pd.DataFrame({'key':['A','B','D'],'value':[4,5,6]})
pd.merge(df1,df2,on='key',how='inner')
```

```python
pd.merge(df1,df2,on='key',how='outer')
```

```python
pd.merge(df1,df2,on='key',how='left')
```

```python
pd.merge(df1,df2,on='key',how='right')
```
