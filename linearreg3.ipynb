```python
import sqlite3
```

```python
connection=sqlite3.connect('student')
cursor=connection.cursor()
```

```python
## create a table
cursor.execute('''
               create table student(
                   id integer,
                   name text,
                   department text)
              ''')
              
```

```python
script='''
insert into student(name,id,department)
VALUES('rohith',19,'CSE');
insert into student(name,id,department)
VALUES('ro',118,'CSE');
insert into student(name,id,department)
VALUES('rohi',190,'CSE');
'''
cursor.executescript(script)
```

```python
connection.commit()
```

```python
cursor.execute('select*from student')
rows=cursor.fetchall()
for row in rows:
    print(row)
```
