```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[6,7,4,8,9]
## create a line plot
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('line plot')
plt.grid=True
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color='black',linestyle='--',linewidth=10,markersize="6")
```

```python
x=[1,2,3,4,5]
y=[6,7,4,8,9]
y2=[1,5,6,8,9]
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.plot(y2,y,color='black')
plt.title('plot 1')
plt.subplot(1,2,2)
plt.plot(x,y,color='blue')
plt.title('plot 2')
```

```python
## bar plot
categories=['a','s','d','f','g']
values=[1,2,3,7,9]
plt.bar(categories,values,color='blue')
```

```python
## histograms
data=[1,1,2,2,2,3,3,3,4,4,4,4,4,5,6,6,6,6]
plt.hist(data,bins=6,color='blue',edgecolor='black')
```

```python
## pie chart
labels=['india','america','japan','china','russia']
population=[40,10,5,30,15]
plt.pie(population,labels=labels,shadow=True)
```
